```python
# import Libraries

import openai
import langchain
import pinecone
from langchain.document_loaders import PyPDFDirectoryLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.vectorstores import Pinecone
from langchain.llms import OpenAI
```

```python
import os
```

```python
## Lets read the document
def read_doc(directory):
    file_loader=PyPDFDirectoryLoader(directory)
    documents=file_loader.load()
    return documents
```

```python
doc=read_doc('documents/')
doc
```

```python
len(doc)
```

```python
## Divide the docs into chunks
def chunk_data(docs, chunk_size=800, chunk_overlap=50):
    text_splitter=RecursiveCharacterTextSplitter(chunk_size=chunk_size,chunk_overlap=chunk_overlap)
    doc=text_splitter.split_documents(docs)
    return docs
```

```python
documents=chunk_data(docs=doc)
documents
```

```python
len(documents)
```

```python
## Word embeddings with OpenAI
embeddings=OpenAIEmbeddings(api_key=os.environ['OPENAI_API_KEY'])
embeddings
```

```python
vectors=embeddings.embed_query("How are you?")
vectors
```

```python
len(vectors)
```

```python
## Vector search DB in Pinecone
pinecone.init(
    api_key=os.environ['PINECONE_DB_API_KEY'],
    environment='gcp-starter'
)
index_name='langchainvector'
```

```python
index=Pinecone.from_documents(doc, embeddings, index_name=index_name)
```

```python
## Cosine Similarity Retrieval from  Pinecone vectorDB
def retrieve_query(query,k=2):
    matching_results=index.similarity_search(query,k=2)
    return matching_results
```

```python
from langchain.chains.question_answering import load_qa_chain
from langchain import OpenAI
```

```python
llm=OpenAI(model_name='text-davinci-003',temperature=0.5)
chain=load_qa_chain(llm,chain_type='stuff')
```

```python
## Search answers from Pinecone vectorDB
def retrieve_answers(query):
    doc_search=retrieve_query(query)
    print(doc_search)
    response=chain.run(input_documents=doc_search,question=query)
    return response
```

```python
our_query = 'What is a perceptron?'
answer = retrieve_answers(our_query)
print(answer)
```
